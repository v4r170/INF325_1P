# Preprocesamiento en Python

```python
import numpy as np
from scipy import stats

import pandas as pd 
df = pd.read_csv("feminicidio2020.csv")

pd.set_option('display.max_rows',None)
#df.head()
#print(df)
df
```

IMPUTACION El dataset tiene campos vacios (NAN), como los datos son cualitativos, se procede a calcular la moda para cada columna

```python
moda=stats.mode(df)
print(moda)
print(moda[0][0])
m=moda[0][0];
```

```python
df["area"]=df["area"].replace(np.nan,m[1])
df["relacion"]=df["relacion"].replace(np.nan,m[2])
df["hijos"]=df["hijos"].replace(np.nan,m[3])
df["victima"]=df["victima"].replace(np.nan,m[4])
df["denunciado"]=df["denunciado"].replace(np.nan,m[5])
df["lugar"]=df["lugar"].replace(np.nan,m[6])
df["causa"]=df["causa"].replace(np.nan,m[7])
df
```

DISCRETIZAR, hijos 1, no hijos 0, valores cuantitativos

```python
df["hijos"]=df["hijos"].replace("Si",1)
df["hijos"]=df["hijos"].replace("No",0)

df
```

CATEGORIZAR, Rural=R, Urbana=U, Valores Cualitativos

```python
df["area"]=df["area"].replace("Rural","R")
df["area"]=df["area"].replace("Urbana","U")

df
```
